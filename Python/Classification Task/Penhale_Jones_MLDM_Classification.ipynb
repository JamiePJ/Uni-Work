## Washington Conclusions
<br>
The Washington team had 3 names during the 2019-2022 seasons ("Washington Redskins" in 2019, "Washington Football Team" in 2020 and 2021, then "Washington Commanders" in 2022). <br>
Possibly due to these name changes, the dataset's author has not included the Washington teams under the "Team A" column  for any of the four 4 seasons. The Washington Team are included in the "Team B" column.<br>
Each game featuring Washington needs to be recreated with Washington as "Team A" so  the classes remain balanced for classification.<br>
<br>
I create a function to:<br>
 - locate each row with "Washington" included in the "Team B" name<br>
 - for each Washington row, create a new row with "Washington Commanders" as "Team A" and the former "Team A" team as "Team B"<br>
 - use the headings to reverse the fixture e.g. Team A passing yards becomes Team B passing yards<br>
 - repeat for all rows containing "Washington" under the "Team B" name
<br>
<br>
This will add all of Washington's games with Washington as "Team A" and also rename the team to "Washington Commanders" for all seasons.

```python
# check size of original dataframe to then check that new rows have added correctly
nfl2019.shape
```

```python
# function to create fixtures with washington as team A and stats reversed
def create_washington_rows(season_data):
    washington_df = season_data[season_data["Team_B"].str.contains("Washington", na=False)]
    new_df = pd.DataFrame()
    
    for i in range(len(washington_df)):
        row = washington_df.iloc[i,]
        
        # set the "Game_Result" for Washington
        if row["Game_Result"] == "W":
            game_result = "L"
        if row["Game_Result"] == "L":
            game_result = "W"
        if row["Game_Result"] == "T":
            game_result = "T"
        # create new row using values from the existing row    
        new_row = {'Week': row["Week"], 
                   'Day': row["Day"], 
                   'Date': row["Date"], 
                   'Game_Result': game_result, 
                   'Team_B': row["Team_A"], 
                   'TeamA_1stD': row["TeamB_1stD"],
                   'TeamA_TotYd': row["TeamB_TotYd"], 
                   'TeamA_PassY': row["TeamB_PassY"], 
                   'TeamA_RushY': row["TeamB_RushY"], 
                   'TeamA_TO': row["TeamB_TO"], 
                   'TeamB_1stD': row["TeamA_1stD"], 
                   'TeamB_TotYd': row["TeamA_TotYd"], 
                   'TeamB_PassY': row["TeamA_PassY"],
                   'TeamB_RushY': row["TeamA_RushY"], 
                   'TeamB_TO': row["TeamA_TO"], 
                   'Offence': np.NaN, 
                   'Defence': np.NaN, 
                   'Sp. Tms': np.NaN, 
                   'Team_A': "Washington Commanders"            
        }
        # create dataframe from new_row dictionary and add to main dataframe
        new_row_df = pd.DataFrame(new_row, index=[0])
        new_df = pd.concat((new_df, new_row_df), ignore_index=True)
    main_df = pd.concat((season_data, new_df), ignore_index=True)
        
    return main_df
```

## Web Scraping Function Descriptions
#### scrape_all_season_fixtures()
Takes the season year as input, opens NFL.com for every week of the regular and postseason for that season. It saves each team's abbreviated name to a list of lists; where the main list is for the whole season and the sublist is for each week of fixtures i.e. Week 1, Week 2, etc. Returns the list of abbreviated team names.

#### convert_team_codes_to_names()
Takes the list of abbreviated team names (name codes) and converts them to the full names for every team. Returns the list of full team names.

#### create_season_fixture_list()
Takes the main season's list of weekly fixtures and subdivides each fixture of two teams into its own list. Returns a list of the whole season, containing lists of weekly fixtures, that contain lists for each fixture i.e. one away_team/home_team pair for every game of the week.

#### create_full_fixture_list()
Takes the season year as input and applies the above 3 functions to scrape the fixtures from NFL.com and create the list of lists for all fixtures in that season. Returns the full fixture list.

```python
# import selenium library and necessary features for web scraping
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# function for scraping the fixtures for every week of one season (including regular & postseason)
def scrape_all_season_fixtures(season):
    # settings for selenium
    cService = webdriver.ChromeService(executable_path=r"C:\Users\jamie\PycharmProjects\MLDM\Assignment\nfl stats\chromedriver-win64\chromedriver.exe")
    driver = webdriver.Chrome(service=cService)
    
    # set length of season for each year
    if season == 2019 or season == 2020:
        length_season = 17
    if season == 2021 or season == 2022:
        length_season = 18
        
    # list for storing each week's fixtures
    fixtures_year = []
    
    # open the NFL.com page for each regular season week and locate the html class for the team abbreviation
    for week in range(1, length_season+1):
        driver.get(f"https://www.nfl.com/schedules/{season}/REG{week}/")
        
        # wait for page to load
        time.sleep(5)
        driver.implicitly_wait(15)
        # find team name abbreviation
        team_names = driver.find_elements(by=By.CLASS_NAME, value="nfl-c-matchup-strip__team-abbreviation")

        # save the team name codes to list of that week's teams
        week_team_codes = []
        for i in range(len(team_names)):
            week_team_codes.append(team_names[i].text)
        # add the week of teams to the year's fixtures
        fixtures_year.append(week_team_codes)
        
    # open the NFL.com page for each postseason game and locate the html class for the team abbreviation
    for post_week in range(1, 5):
        driver.get(f"https://www.nfl.com/schedules/{season}/POST{post_week}/")
        
        # wait for page to load
        time.sleep(5)
        driver.implicitly_wait(5)
        # find team name abbreviation
        team_names = driver.find_elements(by=By.CLASS_NAME, value="nfl-c-matchup-strip__team-abbreviation")
        
        # save the team name codes to list of that week's teams
        week_team_codes = []
        for i in range(len(team_names)):
            week_team_codes.append(team_names[i].text)
            
        # add the week of teams to the year's fixtures
        fixtures_year.append(week_team_codes)
        
    # close the Chrome window
    driver.quit()
    return fixtures_year
```

```python
# function to convert the team name codes to full names
def convert_team_codes_to_names(fixture_team_codes):
    # dictionary for replacing each abbreviated team name with the full team name
    replace_names = {
        "ARI":"Arizona Cardinals",
        "ATL":"Atlanta Falcons",
        "BAL":"Baltimore Ravens",
        "BUF":"Buffalo Bills",
        "CAR":"Carolina Panthers",
        "CHI":"Chicago Bears",
        "CIN":"Cincinnati Bengals",
        "CLE":"Cleveland Browns",
        "DAL":"Dallas Cowboys",
        "DEN":"Denver Broncos",
        "DET":"Detroit Lions",
        "GB":"Green Bay Packers",
        "HOU":"Houston Texans",
        "IND":"Indianapolis Colts",
        "JAC":"Jacksonville Jaguars",
        "JAX":"Jacksonville Jaguars",
        "KC":"Kansas City Chiefs",
        "LA": "Los Angeles Chargers",
        "LAC":"Los Angeles Chargers",
        "LAR":"Los Angeles Rams",
        "LV":"Las Vegas Raiders",
        "MIA":"Miami Dolphins",
        "MIN":"Minnesota Vikings",
        "NE":"New England Patriots",
        "NO":"New Orleans Saints",
        "NYG":"New York Giants",
        "NYJ":"New York Jets",
        "OAK":"Las Vegas Raiders",
        "PHI":"Philadelphia Eagles",
        "PIT":"Pittsburgh Steelers",
        "SEA":"Seattle Seahawks",
        "SF":"San Francisco 49ers",
        "TB":"Tampa Bay Buccaneers",
        "TEN":"Tennessee Titans",
        "WAS":"Washington Commanders"
    }
    # list for saving full names
    fixtures_full_names = []
    
    # iterate through every week of fixtures and convert the team codes to the full names
    for i in range(len(fixture_team_codes)):
        fixtures_week_names = [replace_names.get(x) for x in fixture_team_codes[i]]
        fixtures_full_names.append(fixtures_week_names)
        fixtures_week_names = []
    return fixtures_full_names
```

```python
# function to convert the full team names list to a list for each week of the season, 
# containing lists for each game of that week in format [away_team, home_team]
def create_season_fixture_list(fixtures_full_team_names):
    season_fixtures = []
    for week in fixtures_full_team_names:
        week_fixtures = []
        fixture = []

        for i in range(0, len(week), 2):
            fixture.append(week[i])
            fixture.append(week[i+1])
            week_fixtures.append(fixture)
            fixture = []
        season_fixtures.append(week_fixtures)
    return season_fixtures
```

```python
# function that combines the fixture functions to take a year as input and 
# output the list of fixtures for every week in that year's season
def create_full_fixture_list(year):
    fixture_list = scrape_all_season_fixtures(year)
    fixture_list = convert_team_codes_to_names(fixture_list)
    fixture_list = create_season_fixture_list(fixture_list)
    return fixture_list
```

## Home/Away Function Descriptions
#### assign_home_away_teams()
Finds the home and away team for one week's worth of games. Takes DataFrame of team statistics and the fixtures list (both for one week only) and checks whether Team_A is the home team. Assigns a binary value in column "Team_A_Home" indicating whether Team_A is home or away. Home = 1, Away = 0.<br>
Returns a new DataFrame.
#### assign_home_away_teams_all()
Takes DataFrame of a full season's statistics and the fixtures list for the full season, then applies the <b>assign_home_away_teams()</b> function to every week of the season. 
Returns a new DataFrame with the home/away flag included.

```python
# function to find if Team_A is home team for one week of fixtures
def assign_home_away_teams(team_stats, fixture_list):
    for i in range(len(team_stats)):
        row_a = team_stats.iloc[i,]
        team_a = row_a["Team_A"]
        team_b = row_a["Team_B"]
        if any(item[1] == team_a for item in fixture_list):
            # team_a are home team
            team_stats.loc[i, "Team_A_Home"] = 1
        else:
            # team_a are away team
            team_stats.loc[i, "Team_A_Home"] = 0
    return team_stats
```

```python
# function to find if Team_A is home team for every week of fixtures in a season
def assign_home_away_teams_all(year_stats, fixture_list):
    new_df = pd.DataFrame()
    # for each week in the list of all weeks fixtures
    for j in range(len(fixture_list)):
        week = year_stats[year_stats["Week"]==j+1]
        week.reset_index(drop=True, inplace=True)
        week_home_away_df = assign_home_away_teams(week, fixture_list[j])
        new_df = pd.concat([new_df, week_home_away_df], ignore_index=True)
    return new_df
```

#### Combined Function
Function takes the season's statistics DataFrame and season year as input, applies all the above functions to output a new DataFrame that includes the flag to indicate if Team_A was at home for every game in the season. 

```python
# function for updating dataframe with the home/away flag
def add_home_flag_to_season_stats(season_data, season_year):
    fixtures_year = create_full_fixture_list(season_year)
    updated_df = assign_home_away_teams_all(season_data, fixtures_year)
    return updated_df
```

### Rolling Averages
Works by taking the mean of a specified previous number of values. I take each Team_A's previous 3 matches and calculate the mean of each statistic, then add this to "rolling average" column for that statistic.
<br>
<br>
Group DataFrame by each team , then sort by year and week to keep time-series order. Three week rolling mean is calculated for the statistic, using closed="left" so current week is not included in each rolling calculation.

```python
# group the dataframe by the "Team_A" column
grouped_master_stats = master_stats.groupby("Team_A")
```

```python
# get one team's group from the dataframe
san_fran = grouped_master_stats.get_group("San Francisco 49ers")
```

```python
# sort the grouped dataframe into year and week order
san_fran_sorted = san_fran.sort_values(by=['Year', 'Week'])
san_fran_sorted
```

```python
# calculate the rolling average for one match statistic
san_fran_rolling_1stD = san_fran_sorted["TeamA_1stD"].rolling(3,closed="left").mean()
san_fran_rolling_1stD
```

Create function for calculating rolling averages for all Team_A match statistics. Rolling average assigns NaN to the first 3 rows as it requires 3 full weeks to calculate the average. Replace these NaN with 0. 

```python
# create list of new rolling average columns
stats_cols = ['TeamA_1stD', 'TeamA_TotYd', 'TeamA_PassY',
       'TeamA_RushY', 'TeamA_TO']
new_stats_cols = [f"{stat}_rolling" for stat in stats_cols]
new_stats_cols
```

```python
# function to calculate 3-week rolling average for each Team_A statistic
def calc_rolling_avg(team, stats, new_stats):
    # sort team group by year and week
    team_sorted = team.sort_values(by=['Year', 'Week'])
    rolling_avg_cols = team_sorted[stats].rolling(3,closed="left").mean()
    team_sorted[new_stats] = rolling_avg_cols
    team_sorted.fillna(0, inplace=True)
    return team_sorted
```

```python
# test function on san francisco only
san_fran_rolling = calc_rolling_avg(san_fran, stats_cols, new_stats_cols)
san_fran_rolling
```

```python
# apply rolling averages function to full dataframe
master_stats_rolling = master_stats.groupby("Team_A").apply(lambda x: calc_rolling_avg(x, stats_cols, new_stats_cols))
```
